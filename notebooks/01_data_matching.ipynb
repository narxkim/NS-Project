```python
from pathlib import Path
import shutil
```

```python
IMAGE_DIR = Path("../data/raw/images")
JSON_DIR = Path("../data/raw/annotations")

print("이미지 폴더 존재:", IMAGE_DIR.exists())
print("JSON 폴더 존재:", JSON_DIR.exists())
```

```python
image_files = {
    f.stem: f
    for f in IMAGE_DIR.iterdir()
    if f.suffix.lower() in [".jpg", ".jpeg", ".png"]
}

json_files = {
    f.stem: f
    for f in JSON_DIR.iterdir()
    if f.suffix.lower() == ".json"
}

matched_names = sorted(set(image_files.keys()) & set(json_files.keys()))
unlabeled_names = sorted(set(image_files.keys()) - set(json_files.keys()))
orphan_json_names = sorted(set(json_files.keys()) - set(image_files.keys()))

print("전체 이미지 수:", len(image_files))
print("전체 JSON 수:", len(json_files))
print("매칭 성공:", len(matched_names))
print("라벨 없는 이미지:", len(unlabeled_names))
print("이미지 없는 JSON:", len(orphan_json_names))
```


---
### 라벨링 완료 / 미완료 데이터 분리

```python
# 라벨링 완료 데이터 저장 폴더
LABELED_IMAGE_DIR = Path("../data/labeled/images")
LABELED_JSON_DIR = Path("../data/labeled/annotations")

# 라벨링 미완료 데이터 저장 폴더
UNLABELED_IMAGE_DIR = Path("../data/unlabeled/images")

# 폴더가 없으면 자동 생성
LABELED_IMAGE_DIR.mkdir(parents=True, exist_ok=True)
LABELED_JSON_DIR.mkdir(parents=True, exist_ok=True)
UNLABELED_IMAGE_DIR.mkdir(parents=True, exist_ok=True)
```


---
### 라벨링 완료 데이터 복사
- 이미지와 JSON이 모두 존재하는 경우 -> 학습 가능한 데이터

```python
for name in matched_names:
    shutil.copy2(image_files[name], LABELED_IMAGE_DIR / image_files[name].name)

    shutil.copy2(json_files[name], LABELED_JSON_DIR / json_files[name].name)
```


---
### 라벨링 미완료 데이터 복사
- 이미지는 있지만 JSON이 없는 경우 -> 추후 라벨링 또는 자동 라벨링 대상

```python
for name in unlabeled_names:
    shutil.copy2(image_files[name], UNLABELED_IMAGE_DIR / image_files[name].name)
```



---
### 분리 결과 확인

```python
print("=== 데이터 분리 완료 ===")

print(
    "라벨링 완료 이미지 :",
    len(list(LABELED_IMAGE_DIR.iterdir()))
)

print(
    "라벨링 완료 JSON :",
    len(list(LABELED_JSON_DIR.iterdir()))
)

print(
    "라벨링 미완료 이미지 :",
    len(list(UNLABELED_IMAGE_DIR.iterdir()))
)
```
