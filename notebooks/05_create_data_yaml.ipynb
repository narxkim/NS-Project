```python
from pathlib import Path
import yaml
```

```python
# 프로젝트 기준 data 폴더 경로

DATA_DIR = Path("../data").resolve()

print(DATA_DIR)
```

```python
# YOLO 데이터 설정

data_yaml = {
    "path": str(DATA_DIR),
    "train": "train/images",
    "val": "valid/images",
    "test": "test/images",
    "names": {
        0: "user_character",
        1: "user_id"
    }
}
```

```python
# data.yaml 저장

yaml_path = DATA_DIR / "data.yaml"

with open(yaml_path, "w", encoding="utf-8") as f:
    yaml.dump(
        data_yaml,
        f,
        allow_unicode=True,
        sort_keys=False
    )

print("data.yaml 생성 완료")
print(yaml_path)
```

```python
# 생성된 data.yaml 확인

with open(yaml_path, "r", encoding="utf-8") as f:
    print(f.read())
```
