```python
from pathlib import Path
import shutil

from sklearn.model_selection import train_test_split
```

```python
# 원본 데이터 경로

IMAGE_DIR = Path("../data/labeled/images")
LABEL_DIR = Path("../data/labeled/labels")
```

```python
# 이미지 파일 목록 수집

image_files = sorted(IMAGE_DIR.glob("*.jpg"))

print("이미지 수 : ", len(image_files))
```

```python
# 라벨 txt가 존재하는 이미지만 사용

valid_image_files = []

for img_path in image_files:
    label_path = LABEL_DIR / f"{img_path.stem}.txt"

    if label_path.exists():
        valid_image_files.append(img_path)

print("라벨 매칭 이미지 수 : ", len(valid_image_files))
```

```python
# 현재 경로 확인
print("현재 작업 위치:", Path.cwd())

print("이미지 폴더 존재:", IMAGE_DIR.exists())
print("라벨 폴더 존재:", LABEL_DIR.exists())

print("이미지 샘플:", list(IMAGE_DIR.glob("*"))[:3])
print("라벨 샘플:", list(LABEL_DIR.glob("*"))[:3])
```

```python
# train 70%, temp 30% 분리

train_files, temp_files = train_test_split(
    valid_image_files,
    test_size=0.3,
    random_state=42
)

print("Train :", len(train_files))
print("Temp :", len(temp_files))
```

```python
# valid 15%, test 15% 분리

valid_files, test_files = train_test_split(
    temp_files,
    test_size=0.5,
    random_state=42
)

print("Valid :", len(valid_files))
print("Test :", len(test_files))
```

```python
# 기존 train/valid/test 폴더 삭제

for split in ["train", "valid", "test"]:
    split_dir = Path(f"../data/{split}")

    if split_dir.exists():
        shutil.rmtree(split_dir)

print("기존 split 폴더 삭제 완료")
```

```python
# YOLO 학습용 폴더 생성
 
for split in ["train", "valid", "test"]:
    Path(f"../data/{split}/images").mkdir(parents=True, exist_ok=True)
    Path(f"../data/{split}/labels").mkdir(parents=True, exist_ok=True)
```

```python
# 이미지와 라벨을 함께 복사하는 함수

def copy_dataset(file_list, split_name):
    for img_path in file_list:
        label_path = LABEL_DIR / f"{img_path.stem}.txt"

        shutil.copy2(img_path, Path(f"../data/{split_name}/images/{img_path.name}"))

        shutil.copy2(label_path, Path(f"../data/{split_name}/labels/{label_path.name}"))
```

```python
# 데이터 복사 실행

copy_dataset(train_files, "train")
copy_dataset(valid_files, "valid")
copy_dataset(test_files, "test")

print("데이터셋 분리 완료")
```

```python
# 분리 결과 확인

for split in ["train", "valid", "test"]:
    image_count = len(list(Path(f"../data/{split}/images").glob("*")))
    label_count = len(list(Path(f"../data/{split}/labels").glob("*")))

    print(split, image_count, label_count)
```
